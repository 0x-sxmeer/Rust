```python
data=[[20,34,47,58],
      [67,89,10,15],
      [78,90,12,18]]

data1=[[37,39,36,28],
       [28,34,36,27],
       [29,36,35,]]

pd.DataFrame(data+data1)

ln=pd.DataFrame(data+data1)
ln=ln.rename(columns={0:"a",1:"b",2:"c",3:"d"})
print(ln[0:3])
ln.mean()
print(ln.sum())
print(ln.describe())
print(ln.info())
```

```python
df=pd.DataFrame({"Name":["ron","tommy","aryan","sona","mona"],
                "age":[18,20,19,20,40],
                "marks":[98,89,79,55,68]})
df.index=["r1","r2","r3","r4","r5"]
print(df)
ln=df.groupby(["age"])["marks"].sum()
ln=df.groupby(["age"]).agg({"marks":["sum","mean","median"]})
print(ln)
```
